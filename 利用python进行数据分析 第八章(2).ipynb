# 索引上的合并

```python
left1 = DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                  'value': range(6)})
```

```python
right1 = DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')    #这样可以取到并集
```

```python
lefth = DataFrame({'key1': ['Ohio', 'Ohio', 'Ohio',       #层次化索引的数据稍微复杂一些
                           'Nevada', 'Nevada'],
                  'key2': [2000, 2001, 2002, 2001, 2002],
                  'data': np.arange(5.)})
```

```python
righth = DataFrame(np.arange(12).reshape(6,2),
                  index=[['Nevada', 'Nevada', 'Ohio', 'Ohio', 'Ohio', 'Ohio'],
                        [2001, 2000, 2000, 2000, 2001, 2002]],
                  columns=['event1', 'event2'])
```

```python
lefth
```

```python
righth
```

```python
pd.merge(lefth, righth, left_on=['key1', 'key2'], right_index=True)   #默认取交集
```

```python
pd.merge(lefth, righth, left_on=['key1', 'key2'], right_index=True, how='outer')   #取并集
```

```python
left2 = DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                 index=['a', 'c', 'e'],
                 columns=['Ohio', 'Nevada'])
```

```python
right2 = DataFrame([[7., 8.],[9., 10.], [11., 12.], [13, 14]],
                  index=['b', 'c', 'd', 'e'],
                  columns=['Messouri', 'Alabama'])
```

```python
left2
```

```python
right2
```

```python
pd.merge(left2, right2, how='outer', left_index=True, right_index=True)
```

```python
left2.join(right2, how='outer')    #join默认使用左连接，保留左边表的行索引
```

```python
left1.join(right1, on='key')
```

```python
another = DataFrame([[7., 8.], [9., 10.], [11., 12.], [16., 17.]],
                   index=['a','c','e','f'],
                   columns=['New York', 'Oregon'])
```

```python
another
```

```python
left2.join([right2, another])    #可以向join传入一组DataFrame
```

```python
left2.join([right2, another], how='outer')
```

# 合并重叠数据

```python
a = Series([np.nan, 2.5, np.nan, 3.5, 4.5, np.nan],
          index=['f', 'e', 'd', 'c', 'b', 'a'])
```

```python
b = Series(np.arange(len(a), dtype=np.float64),
          index=['f', 'e', 'd', 'c', 'b', 'a'])
```

```python
b[-1] =np.nan
```

```python
a
```

```python
b
```

```python
np.where(pd.isnull(a), b, a)
```

```python
b[:-2].combine_first(a[2:])
```

```python
df1 = DataFrame({'a': [1., np.nan, 5., np.nan],
                'b': [np.nan, 2., np.nan, 6],
                'c': range(2, 18, 4)})
```

```python
s2 = Series([4, 5, 6], index=['c', 'd', 'e'])
```

```python
data2 =pd.concat([s1, s2], keys=['one', 'two'])
```

```python
data2    
```

# 将款格式旋转为长格式

```python
df = DataFrame({'key': ['foo', 'bar', 'baz'],
               'A': [1, 2, 3],
               'B': [4, 5, 6],
               'C': [7, 8, 9]})
```

```python
df
```

```python
melted = pd.melt(df, ['key'])   #将key作为唯一的分组指标，其他的列是数据值
```

```python
melted
```

```python
reshaped = melted.pivot('key', 'variable', 'value')
```

```python
reshaped
```

```python
reshaped.reset_index()    #将被索引化的key列移回数据列
```

```python
pd.melt(df, id_vars=['key'], value_vars=['A', 'B'])   #仅仅指定A、B列为值的列
```

```python
pd.melt(df, value_vars=['A', 'B', 'C'])   #可以不指定分组指标
```

```python
pd.melt(df, value_vars=['key', 'A', 'B'])
```
